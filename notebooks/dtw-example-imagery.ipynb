# Notebook to create example images for DTW

This notebook is used to create example images for explaining DTW in the master thesis.
The code is taken from the [dtaidistance documentation](https://dtaidistance.readthedocs.io/en/latest/usage/dtw.html) and slightly adapted

```python
from dtaidistance import dtw
from dtaidistance import dtw_visualisation as dtwvis
import random
import numpy as np
```

```python
x = np.arange(0, 20, .5)
s1 = np.sin(x)
s2 = np.cos(x - 5)
random.seed(1)
for idx in range(len(s2)):
    if random.random() < 0.05:
        s2[idx] += (random.random() - 0.5) / 2
d, paths = dtw.warping_paths(s1, s2)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(s1, s2, paths, best_path)

d, paths = dtw.warping_paths(s1, s2, window=12)
best_path = dtw.best_path(paths)
dtwvis.plot_warpingpaths(s1, s2, paths, best_path)
```
