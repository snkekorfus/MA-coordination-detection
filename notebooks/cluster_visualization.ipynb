# Generate example imagery for clustering

This notebook is used to create images that help explaining the concept of clustering in the master thesis

```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
x_1 = np.random.normal(2, 1.5, 500)
y_1 = np.random.normal(2, 1.5, 500)
c_1 = np.full((500,), 1)
x_2 = np.random.normal(8, 1.5, 500)
y_2 = np.random.normal(8, 1.5, 500)
c_2 = np.full((500,), 2)
x = np.append(x_1, x_2)
y = np.append(y_1, y_2)
c = np.append(c_1, c_2)
```

```python

plot.figure(figsize=(8, 5), dpi=100)
plot.scatter(x, y)
ax = plot.gca()
ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])
plot.savefig("./cluster_clean.png", dpi=300)
```

```python
plot.figure(figsize=(8, 5), dpi=100)
plot.scatter(x, y, c=c)
ax = plot.gca()
ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])
plot.savefig("./cluster_color.png", dpi=300)
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters=2)
clf.fit(np.vstack((x, y)).T)
c = clf.predict(np.vstack((x, y)).T)
```

```python
plot.figure(figsize=(8, 5), dpi=100)
plot.scatter(x, y, c=c)
ax = plot.gca()
ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])
plot.savefig("./cluster_kmeans.png", dpi=300)
```
